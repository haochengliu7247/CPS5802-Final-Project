# About Dataset
## Title: Synthetic Dataset for Age Prediction

Description:

This dataset contains synthetic data designed for predicting age based on various health and lifestyle factors. It includes 3,000 rows with 24 features, each representing different aspects of physical health and lifestyle.

Features:

- Height (cm): The height of the individual in centimeters.
- Weight (kg): The weight of the individual in kilograms.
- Blood Pressure (s/d): Blood pressure (systolic/diastolic) in mmHg.
- Cholesterol Level (mg/dL): Cholesterol level in milligrams per deciliter.
- BMI: Body Mass Index, calculated from height and weight.
- Blood Glucose Level (mg/dL): Blood glucose level in milligrams per deciliter.
- Bone Density (g/cm²): Bone density in grams per square centimeter.
- Vision Sharpness: Vision sharpness on a scale from 0 (blurry) to 100 (perfect).
- Hearing Ability (dB): Hearing ability in decibels.
- Physical Activity Level: Categorized as 'Low', 'Moderate', or 'High'.
- Smoking Status: Categorical values including 'Never', 'Former', and 'Current'.
- Alcohol Consumption: Frequency of alcohol consumption.
- Diet: Type of diet, categorized as 'Balanced', 'High Protein', 'Low Carb', etc.
- Chronic Diseases: Presence of chronic diseases (e.g., diabetes, hypertension).
- Medication Use: Usage of medication.
- Family History: Presence of family history of age-related conditions.
- Cognitive Function: Self-reported cognitive function on a scale from 0 (poor) to 100 (excellent).
- Mental Health Status: Self-reported mental health status on a scale from 0 (poor) to 100 (excellent).
- Sleep Patterns: Average number of sleep hours per night.
- Stress Levels: Self-reported stress levels on a scale from 0 (low) to 100 (high).
- Pollution Exposure: Exposure to pollution measured in arbitrary units.
- Sun Exposure: Average sun exposure in hours per week.
- Education Level: Highest level of education attained.
- Income Level: Annual income in USD.
- Age (years): The target variable representing the age of the individual.

## Step 2: Data Preprocessing + Multi-Model Age Prediction

This step trains the age-prediction models. It starts with the original baseline pipeline, then compares multiple regression models using one shared preprocessing workflow.

   baseline pipeline  

Important clarification: this step is ordinary machine learning regression, not Qwen/LLM and not local rule-based recommendation.

````python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

try:
    from xgboost import XGBRegressor
    XGBOOST_AVAILABLE = True
    XGBOOST_IMPORT_ERROR = None
except Exception as exc:
    XGBRegressor = None
    XGBOOST_AVAILABLE = False
    XGBOOST_IMPORT_ERROR = exc


class TabularPreprocessor(object):
    """Pandas-based preprocessing that works with older sklearn versions.
      Pandas   sklearn """

    def fit(self, X, y=None):
        X_frame = pd.DataFrame(X).copy()
        self.columns_ = list(X_frame.columns)
        self.numeric_features_ = X_frame.select_dtypes(include=[np.number]).columns.tolist()
        self.categorical_features_ = [
            column for column in self.columns_
            if column not in self.numeric_features_
        ]
        self.numeric_fill_values_ = {}
        for column in self.numeric_features_:
            median_value = pd.to_numeric(X_frame[column], errors='coerce').median()
            if pd.isna(median_value):
                median_value = 0
            self.numeric_fill_values_[column] = median_value

        prepared = self._prepare_frame(X_frame)
        encoded = pd.get_dummies(prepared, columns=self.categorical_features_)
        self.feature_names_ = list(encoded.columns)
        return self

    def _prepare_frame(self, X):
        X_frame = pd.DataFrame(X).copy()
        for column in self.columns_:
            if column not in X_frame.columns:
                X_frame[column] = np.nan

        X_frame = X_frame[self.columns_].copy()
        for column in self.numeric_features_:
            X_frame[column] = pd.to_numeric(X_frame[column], errors='coerce')
            X_frame[column] = X_frame[column].fillna(self.numeric_fill_values_[column])

        for column in self.categorical_features_:
            X_frame[column] = X_frame[column].fillna('Missing').astype(str)

        return X_frame

    def transform(self, X):
        prepared = self._prepare_frame(X)
        encoded = pd.get_dummies(prepared, columns=self.categorical_features_)
        encoded = encoded.reindex(columns=self.feature_names_, fill_value=0)
        return encoded.values.astype(float)

    def get_feature_names(self):
        return list(self.feature_names_)

    def get_feature_names_out(self):
        return np.asarray(self.feature_names_)


class TabularRegressorPipeline(object):
    """Small pipeline replacement for environments without ColumnTransformer.
      ColumnTransformer  pipeline """

    def __init__(self, model=None, preprocessor=None):
        self.model = model if model is not None else make_regression_model()
        self.preprocessor = preprocessor if preprocessor is not None else TabularPreprocessor()
        self.named_steps = {
            'preprocessor': self.preprocessor,
            'model': self.model,
        }

    def fit(self, X, y):
        self.preprocessor.fit(X, y)
        X_transformed = self.preprocessor.transform(X)
        self.model.fit(X_transformed, y)
        return self

    def predict(self, X):
        X_transformed = self.preprocessor.transform(X)
        return self.model.predict(X_transformed)


def make_regression_model():
    if XGBOOST_AVAILABLE:
        return XGBRegressor(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=5,
            random_state=42,
            objective='reg:squarederror',
            n_jobs=-1,
        )
    return RandomForestRegressor(
        n_estimators=120,
        random_state=42,
        n_jobs=-1,
    )


def make_age_pipeline():
    return TabularRegressorPipeline(model=make_regression_model())


if XGBOOST_AVAILABLE:
    print('Using XGBoost for age prediction.')
else:
    print('xgboost is not installed; using RandomForestRegressor instead.')
    print(f'xgboost import message: {XGBOOST_IMPORT_ERROR}')
````

### 2.2 Multi-Model Comparison and Best Model Selection

This section compares Linear Regression, Random Forest, XGBoost, and MLP Regressor with the same train/validation split and preprocessing workflow. The best model is selected by the lowest validation MAE.

 / Linear Regression Random Forest XGBoost  MLP Regressor  MAE  

````python
# Model Comparison: shared preprocessing, unified validation split, and multiple regressors.
# English note: see the surrounding code for this project workflow detail.
import os
import re
import json
import warnings
from collections import OrderedDict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler

try:
    from sklearn.exceptions import ConvergenceWarning
    warnings.filterwarnings('ignore', category=ConvergenceWarning)
except Exception:
    pass
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)

if 'display' not in globals():
    try:
        from IPython.display import display
    except ImportError:
        def display(obj):
            print(obj)

RANDOM_STATE = 42
TARGET_COLUMN = 'Age (years)'

train_data = pd.read_csv('Train.csv')
test_data = pd.read_csv('Test.csv')
X_all = train_data.drop(TARGET_COLUMN, axis=1)
y_all = train_data[TARGET_COLUMN].astype(float)


class AgeFeaturePreprocessor(object):
    """Shared preprocessing for all age-prediction models.
      """

    def __init__(self, drop_original_blood_pressure=True, drop_first=True):
        self.drop_original_blood_pressure = drop_original_blood_pressure
        self.drop_first = drop_first

    def _engineer_features(self, X):
        X_frame = pd.DataFrame(X).copy()
        bp_column = 'Blood Pressure (s/d)'
        if bp_column in X_frame.columns:
            extracted = X_frame[bp_column].astype(str).str.extract(
                r'^\s*(\d+(?:\.\d+)?)\s*/\s*(\d+(?:\.\d+)?)\s*$',
                expand=True,
            )
            X_frame['Systolic Blood Pressure'] = pd.to_numeric(extracted.iloc[:, 0], errors='coerce')
            X_frame['Diastolic Blood Pressure'] = pd.to_numeric(extracted.iloc[:, 1], errors='coerce')
            if self.drop_original_blood_pressure:
                X_frame = X_frame.drop(bp_column, axis=1)
        return X_frame

    def fit(self, X, y=None):
        X_frame = self._engineer_features(X)
        self.columns_ = list(X_frame.columns)
        self.numeric_features_ = X_frame.select_dtypes(include=[np.number]).columns.tolist()
        self.categorical_features_ = [
            column for column in self.columns_
            if column not in self.numeric_features_
        ]
        self.numeric_fill_values_ = {}
        for column in self.numeric_features_:
            median_value = pd.to_numeric(X_frame[column], errors='coerce').median()
            if pd.isna(median_value):
                median_value = 0.0
            self.numeric_fill_values_[column] = float(median_value)

        prepared = self._prepare_frame(X_frame)
        encoded = pd.get_dummies(prepared, columns=self.categorical_features_, drop_first=self.drop_first)
        self.feature_names_ = list(encoded.columns)
        return self

    def _prepare_frame(self, X_engineered):
        X_frame = pd.DataFrame(X_engineered).copy()
        for column in self.columns_:
            if column not in X_frame.columns:
                X_frame[column] = np.nan
        X_frame = X_frame[self.columns_].copy()

        for column in self.numeric_features_:
            X_frame[column] = pd.to_numeric(X_frame[column], errors='coerce')
            X_frame[column] = X_frame[column].fillna(self.numeric_fill_values_[column])

        for column in self.categorical_features_:
            X_frame[column] = X_frame[column].fillna('Missing').astype(str)

        return X_frame

    def transform(self, X):
        X_engineered = self._engineer_features(X)
        prepared = self._prepare_frame(X_engineered)
        encoded = pd.get_dummies(prepared, columns=self.categorical_features_, drop_first=self.drop_first)
        encoded = encoded.reindex(columns=self.feature_names_, fill_value=0)
        return encoded.values.astype(float)

    def fit_transform(self, X, y=None):
        self.fit(X, y)
        return self.transform(X)

    def get_feature_names(self):
        return list(self.feature_names_)


class AgeModelWorkflow(object):
    """Small sklearn-like workflow: preprocess -> scale -> model.
      sklearn   ->  ->  """

    def __init__(self, model, scale_features=True):
        self.model = model
        self.preprocessor = AgeFeaturePreprocessor()
        self.scale_features = scale_features
        self.scaler = StandardScaler() if scale_features else None

    def fit(self, X, y):
        self.target_min_ = float(np.min(y))
        self.target_max_ = float(np.max(y))
        X_processed = self.preprocessor.fit_transform(X, y)
        if self.scaler is not None:
            X_processed = self.scaler.fit_transform(X_processed)
        self.model.fit(X_processed, y)
        return self

    def predict(self, X):
        X_processed = self.preprocessor.transform(X)
        if self.scaler is not None:
            X_processed = self.scaler.transform(X_processed)
        predictions = np.asarray(self.model.predict(X_processed)).astype(float)
        if hasattr(self, 'target_min_') and hasattr(self, 'target_max_'):
            predictions = np.clip(predictions, self.target_min_, self.target_max_)
        return predictions

    def get_feature_names(self):
        return self.preprocessor.get_feature_names()


def make_model_factories():
    factories = OrderedDict()
    factories['Linear Regression'] = lambda: LinearRegression()
    factories['Random Forest'] = lambda: RandomForestRegressor(
        n_estimators=220,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        min_samples_leaf=2,
    )

    xgboost_status = 'available'
    try:
        from xgboost import XGBRegressor
        factories['XGBoost'] = lambda: XGBRegressor(
            n_estimators=260,
            learning_rate=0.05,
            max_depth=4,
            subsample=0.9,
            colsample_bytree=0.9,
            objective='reg:squarederror',
            random_state=RANDOM_STATE,
            n_jobs=-1,
            verbosity=0,
        )
    except Exception as exc:
        xgboost_status = 'not available: {}'.format(exc)

    factories['MLP Regressor'] = lambda: MLPRegressor(
        hidden_layer_sizes=(64, 32),
        activation='relu',
        solver='adam',
        alpha=0.0005,
        learning_rate_init=0.001,
        max_iter=900,
        early_stopping=True,
        validation_fraction=0.15,
        random_state=RANDOM_STATE,
    )
    return factories, xgboost_status


model_factories, xgboost_status = make_model_factories()
print('XGBoost status: {}'.format(xgboost_status))

X_train_part, X_valid_part, y_train_part, y_valid_part = train_test_split(
    X_all,
    y_all,
    test_size=0.20,
    random_state=RANDOM_STATE,
)

comparison_rows = []
trained_validation_workflows = OrderedDict()
validation_predictions_by_model = OrderedDict()
model_failure_logs = OrderedDict()

for model_name, factory in model_factories.items():
    print('\nTraining {}...'.format(model_name))
    try:
        workflow = AgeModelWorkflow(factory(), scale_features=True)
        workflow.fit(X_train_part, y_train_part)
        valid_predictions = workflow.predict(X_valid_part)

        mse = mean_squared_error(y_valid_part, valid_predictions)
        rmse = float(np.sqrt(mse))
        mae = mean_absolute_error(y_valid_part, valid_predictions)
        r2 = r2_score(y_valid_part, valid_predictions)

        comparison_rows.append({
            'Model': model_name,
            'MSE': float(mse),
            'RMSE': rmse,
            'MAE': float(mae),
            'R2': float(r2),
            'Validation Rows': int(len(y_valid_part)),
            'Status': 'OK',
        })
        trained_validation_workflows[model_name] = workflow
        validation_predictions_by_model[model_name] = valid_predictions
        print('{} done: MAE={:.3f}, RMSE={:.3f}, R2={:.3f}'.format(model_name, mae, rmse, r2))
    except Exception as exc:
        model_failure_logs[model_name] = str(exc)
        comparison_rows.append({
            'Model': model_name,
            'MSE': np.nan,
            'RMSE': np.nan,
            'MAE': np.nan,
            'R2': np.nan,
            'Validation Rows': int(len(y_valid_part)),
            'Status': 'FAILED: {}'.format(exc),
        })
        print('{} failed: {}'.format(model_name, exc))

model_comparison_df = pd.DataFrame(comparison_rows)
model_comparison_df = model_comparison_df.sort_values(['MAE', 'RMSE'], ascending=[True, True]).reset_index(drop=True)

successful_model_comparison_df = model_comparison_df[model_comparison_df['Status'] == 'OK'].copy()
if len(successful_model_comparison_df) == 0:
    raise RuntimeError('No model finished successfully. Please check the model failure logs.')

best_model_name = successful_model_comparison_df.iloc[0]['Model']
best_model_workflow = trained_validation_workflows[best_model_name]
best_model_metrics = successful_model_comparison_df.iloc[0].to_dict()

print('\nModel comparison table sorted by MAE:')
display(model_comparison_df)

print('\nBest model: {}'.format(best_model_name))
print(
    'The best model is selected because it has the lowest validation MAE ({:.3f} years). '
    'MAE is the primary metric here because it is directly interpretable as average age error in years.'.format(
        float(best_model_metrics['MAE'])
    )
)
````

## Step 3: Best Model Selection + Test Prediction Export

The best model name is produced in Step 2. This section retrains each successful model on the full `Train.csv` dataset and exports `Test.csv` body-age predictions under `model_outputs/`.

 2  `Train.csv`   `Test.csv`  `model_outputs/` 

````python
# Export Prediction CSV Files: retrain each model on full training data and predict Test.csv.
# English note: see the surrounding code for this project workflow detail.
MODEL_OUTPUT_DIR = os.path.abspath('model_outputs')
os.makedirs(MODEL_OUTPUT_DIR, exist_ok=True)

model_output_folder_names = {
    'Linear Regression': 'LinearRegression',
    'Random Forest': 'RandomForest',
    'XGBoost': 'XGBoost',
    'MLP Regressor': 'MLPRegressor',
}

export_summary_rows = []
full_train_model_workflows = OrderedDict()
final_test_predictions_by_model = OrderedDict()

for model_name, factory in model_factories.items():
    folder_name = model_output_folder_names.get(model_name, re.sub(r'[^A-Za-z0-9]+', '', model_name))
    model_dir = os.path.join(MODEL_OUTPUT_DIR, folder_name)
    os.makedirs(model_dir, exist_ok=True)
    output_path = os.path.join(model_dir, 'test_predictions.csv')

    print('\nRetraining {} on the full training data...'.format(model_name))
    try:
        workflow = AgeModelWorkflow(factory(), scale_features=True)
        workflow.fit(X_all, y_all)
        test_predictions = workflow.predict(test_data)
        prediction_output_df = pd.DataFrame({
            'Person Index': np.arange(len(test_data), dtype=int),
            'Predicted Body Age (years)': np.round(test_predictions.astype(float), 2),
        })
        prediction_output_df.to_csv(output_path, index=False)

        full_train_model_workflows[model_name] = workflow
        final_test_predictions_by_model[model_name] = test_predictions
        export_summary_rows.append({
            'Model': model_name,
            'Rows': int(len(prediction_output_df)),
            'CSV Path': output_path,
            'Status': 'saved',
        })
        print('Saved {}'.format(output_path))
    except Exception as exc:
        export_summary_rows.append({
            'Model': model_name,
            'Rows': 0,
            'CSV Path': output_path,
            'Status': 'skipped: {}'.format(exc),
        })
        print('Skipped {} because: {}'.format(model_name, exc))

if 'XGBoost' not in model_factories:
    export_summary_rows.append({
        'Model': 'XGBoost',
        'Rows': 0,
        'CSV Path': os.path.join(MODEL_OUTPUT_DIR, 'XGBoost', 'test_predictions.csv'),
        'Status': 'skipped: {}'.format(xgboost_status),
    })

exported_prediction_files_df = pd.DataFrame(export_summary_rows)
print('\nPrediction CSV export summary:')
display(exported_prediction_files_df)
````

## Step 4: Prediction Explainability

This module explains body-age predictions with a local sensitivity method. It changes one feature at a time around the selected synthetic profile, then shows which factors push the predicted body age higher or lower.


Important clarification: this is not SHAP, not Qwen, and not rule-based recommendation. It explains the age-prediction model.

````python
# English note: see the surrounding code for this project workflow detail.
# English note: see the surrounding code for this project workflow detail.
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor

if 'display' not in globals():
    try:
        from IPython.display import display
    except ImportError:
        def display(obj):
            print(obj)

explainability_train_data = train_data.copy() if 'train_data' in globals() else pd.read_csv('Train.csv')
explainability_test_data = test_data.copy() if 'test_data' in globals() else pd.read_csv('Test.csv')

explainability_X = explainability_train_data.drop('Age (years)', axis=1)
explainability_y = explainability_train_data['Age (years)']
explainability_numeric_features = explainability_X.select_dtypes(include=[np.number]).columns.tolist()
explainability_categorical_features = [
    column for column in explainability_X.columns
    if column not in explainability_numeric_features
]

if 'TabularPreprocessor' not in globals():
    class TabularPreprocessor(object):
        """Pandas-based preprocessing for standalone explainability runs.
          Pandas  """

        def fit(self, X, y=None):
            X_frame = pd.DataFrame(X).copy()
            self.columns_ = list(X_frame.columns)
            self.numeric_features_ = X_frame.select_dtypes(include=[np.number]).columns.tolist()
            self.categorical_features_ = [
                column for column in self.columns_
                if column not in self.numeric_features_
            ]
            self.numeric_fill_values_ = {}
            for column in self.numeric_features_:
                median_value = pd.to_numeric(X_frame[column], errors='coerce').median()
                if pd.isna(median_value):
                    median_value = 0
                self.numeric_fill_values_[column] = median_value

            prepared = self._prepare_frame(X_frame)
            encoded = pd.get_dummies(prepared, columns=self.categorical_features_)
            self.feature_names_ = list(encoded.columns)
            return self

        def _prepare_frame(self, X):
            X_frame = pd.DataFrame(X).copy()
            for column in self.columns_:
                if column not in X_frame.columns:
                    X_frame[column] = np.nan

            X_frame = X_frame[self.columns_].copy()
            for column in self.numeric_features_:
                X_frame[column] = pd.to_numeric(X_frame[column], errors='coerce')
                X_frame[column] = X_frame[column].fillna(self.numeric_fill_values_[column])

            for column in self.categorical_features_:
                X_frame[column] = X_frame[column].fillna('Missing').astype(str)

            return X_frame

        def transform(self, X):
            prepared = self._prepare_frame(X)
            encoded = pd.get_dummies(prepared, columns=self.categorical_features_)
            encoded = encoded.reindex(columns=self.feature_names_, fill_value=0)
            return encoded.values.astype(float)

if 'TabularRegressorPipeline' not in globals():
    class TabularRegressorPipeline(object):
        """Small pipeline replacement for standalone explainability runs.
          pipeline """

        def __init__(self, model=None, preprocessor=None):
            self.model = model if model is not None else RandomForestRegressor(
                n_estimators=120,
                random_state=42,
                n_jobs=-1,
            )
            self.preprocessor = preprocessor if preprocessor is not None else TabularPreprocessor()
            self.named_steps = {
                'preprocessor': self.preprocessor,
                'model': self.model,
            }

        def fit(self, X, y):
            self.preprocessor.fit(X, y)
            X_transformed = self.preprocessor.transform(X)
            self.model.fit(X_transformed, y)
            return self

        def predict(self, X):
            X_transformed = self.preprocessor.transform(X)
            return self.model.predict(X_transformed)

if 'make_age_pipeline' not in globals():
    def make_age_pipeline():
        return TabularRegressorPipeline()


def _existing_pipeline_is_usable(candidate_pipeline, sample_frame):
    """ pipeline  """
    try:
        candidate_pipeline.predict(sample_frame.iloc[[0]])
        return True
    except Exception:
        return False


# Prefer the best full-train workflow exported in Step 3, because later recommendations use the same prediction source.
# English note: see the surrounding code for this project workflow detail.
if (
    'full_train_model_workflows' in globals()
    and 'best_model_name' in globals()
    and best_model_name in full_train_model_workflows
    and _existing_pipeline_is_usable(full_train_model_workflows[best_model_name], explainability_test_data)
):
    explainability_pipeline = full_train_model_workflows[best_model_name]
    explainability_model_source = 'best full-train model workflow: {}'.format(best_model_name)
elif 'pipeline' in globals() and _existing_pipeline_is_usable(pipeline, explainability_test_data):
    explainability_pipeline = pipeline
    explainability_model_source = 'existing baseline notebook prediction pipeline'
else:
    explainability_pipeline = make_age_pipeline()
    explainability_pipeline.fit(explainability_X, explainability_y)
    explainability_model_source = 'new RandomForest-compatible prediction pipeline'

explainability_predictions = explainability_pipeline.predict(explainability_test_data)
explainability_predictions_df = pd.DataFrame({
    'Person Index': np.arange(len(explainability_predictions)),
    'Predicted Body Age (years)': np.round(explainability_predictions, 2),
})

print('Prediction + Explainability model is ready.')
print(f'Model source: {explainability_model_source}')
print(f'Prediction rows: {len(explainability_predictions_df)}')
display(explainability_predictions_df.head())
````

````python
# English note: see the surrounding code for this project workflow detail.
# English note: see the surrounding code for this project workflow detail.

def _is_explain_missing(value):
    """ """
    if value is None:
        return True
    try:
        if pd.isna(value):
            return True
    except (TypeError, ValueError):
        pass
    return str(value).strip().lower() in {'', 'nan', 'none', 'null'}


def _format_explain_value(value):
    """ """
    if _is_explain_missing(value):
        return 'Not provided'
    if isinstance(value, (int, float, np.integer, np.floating)):
        return round(float(value), 3)
    return str(value)


def _make_reference_values():
    """  """
    reference_values = {}
    for feature in explainability_X.columns:
        if feature in explainability_numeric_features:
            median_value = pd.to_numeric(explainability_X[feature], errors='coerce').median()
            if pd.isna(median_value):
                median_value = 0
            reference_values[feature] = median_value
        else:
            non_missing_values = explainability_X[feature].dropna().astype(str)
            if len(non_missing_values) == 0:
                reference_values[feature] = 'Missing'
            else:
                reference_values[feature] = non_missing_values.value_counts().index[0]
    return reference_values


explainability_reference_values = _make_reference_values()


def _make_prediction_reason(predicted_age, base_value, top_contributions_df):
    """ Why this prediction  """
    increasing = (
        top_contributions_df[top_contributions_df['Contribution (years)'] > 0]
        .sort_values('Contribution (years)', ascending=False)
        .head(3)
    )
    decreasing = (
        top_contributions_df[top_contributions_df['Contribution (years)'] < 0]
        .sort_values('Contribution (years)', ascending=True)
        .head(3)
    )

    lines = [
        f'The model predicted Body Age = {predicted_age:.2f} years.',
        f'A typical reference profile is predicted at {base_value:.2f} years.',
        'Each factor below estimates how the current value changes the prediction compared with replacing only that feature with a typical training value.',
    ]
    if len(increasing) > 0:
        lines.append(
            'Main factors increasing predicted body age: '
            + '; '.join(
                f"{row['Feature']} ({row['Value']}) contributed {row['Contribution']}"
                for _, row in increasing.iterrows()
            )
            + '.'
        )
    if len(decreasing) > 0:
        lines.append(
            'Main factors decreasing predicted body age: '
            + '; '.join(
                f"{row['Feature']} ({row['Value']}) contributed {row['Contribution']}"
                for _, row in decreasing.iterrows()
            )
            + '.'
        )
    return '\n'.join(lines)


def explain_body_age_prediction(person_index=0, top_n=12):
    """ """
    if person_index < 0 or person_index >= len(explainability_test_data):
        raise IndexError(f'person_index must be between 0 and {len(explainability_test_data) - 1}.')

    row_frame = explainability_test_data.iloc[[person_index]].copy()
    predicted_age = float(explainability_pipeline.predict(row_frame)[0])

    reference_frame = pd.DataFrame(
        [explainability_reference_values],
        columns=explainability_X.columns,
    )
    base_value = float(explainability_pipeline.predict(reference_frame)[0])

    original_row = row_frame.iloc[0]
    contribution_rows = []
    for feature in explainability_X.columns:
        counterfactual_frame = row_frame.copy()
        counterfactual_frame.loc[:, feature] = explainability_reference_values[feature]
        counterfactual_prediction = float(explainability_pipeline.predict(counterfactual_frame)[0])
        contribution_value = predicted_age - counterfactual_prediction
        contribution_rows.append({
            'Feature': feature,
            'Value': _format_explain_value(original_row.get(feature, 'Not provided')),
            'Reference Value': _format_explain_value(explainability_reference_values.get(feature, 'Not provided')),
            'Contribution (years)': contribution_value,
        })

    contribution_df = pd.DataFrame(contribution_rows)
    contribution_df['Absolute Contribution (years)'] = contribution_df['Contribution (years)'].abs()
    contribution_df = contribution_df.sort_values('Absolute Contribution (years)', ascending=False)
    contribution_df['Contribution (years)'] = contribution_df['Contribution (years)'].round(3)
    contribution_df['Contribution'] = contribution_df['Contribution (years)'].map(lambda value: f'{value:+.3f} years')
    contribution_df['Impact'] = np.where(
        contribution_df['Contribution (years)'] >= 0,
        'Increases predicted body age',
        'Decreases predicted body age',
    )

    columns = ['Feature', 'Value', 'Reference Value', 'Contribution', 'Contribution (years)', 'Impact']
    top_contributions = contribution_df[columns].head(top_n).reset_index(drop=True)

    return {
        'person_index': person_index,
        'predicted_body_age': round(predicted_age, 2),
        'base_value': round(base_value, 2),
        'top_contributions': top_contributions,
        'all_contributions': contribution_df[columns].reset_index(drop=True),
        'explanation_text': _make_prediction_reason(predicted_age, base_value, top_contributions),
    }

print('Explainability functions are ready.')
````

````python
# Final LLM recommendation helper functions.
# English note: see the surrounding code for this project workflow detail.

import os
import json
import re
from pathlib import Path
from urllib import request

import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Define display when the notebook code is executed in a plain Python environment.
# English note: see the surrounding code for this project workflow detail.
if 'display' not in globals():
    try:
        from IPython.display import display
    except ImportError:
        def display(obj):
            print(obj)

def _is_missing(value):
    """ """
    if value is None:
        return True
    try:
        if pd.isna(value):
            return True
    except (TypeError, ValueError):
        pass
    return str(value).strip().lower() in {'', 'nan', 'none', 'null'}


def _clean_value(value, default='Not provided'):
    """ """
    return default if _is_missing(value) else str(value)


def _to_float(value):
    """  None """
    if _is_missing(value):
        return None
    try:
        return float(value)
    except (TypeError, ValueError):
        return None


def _parse_blood_pressure(value):
    """ 151/109  """
    if _is_missing(value):
        return None, None
    match = re.search(r'(\d+(?:\.\d+)?)\s*/\s*(\d+(?:\.\d+)?)', str(value))
    if not match:
        return None, None
    return float(match.group(1)), float(match.group(2))


def _collect_health_signals(row):
    """ """
    signals = []
    bmi = _to_float(row.get('BMI'))
    cholesterol = _to_float(row.get('Cholesterol Level (mg/dL)'))
    glucose = _to_float(row.get('Blood Glucose Level (mg/dL)'))
    stress = _to_float(row.get('Stress Levels'))
    predicted_age = _to_float(row.get('Predicted Body Age (years)'))
    systolic, diastolic = _parse_blood_pressure(row.get('Blood Pressure (s/d)'))

    if predicted_age is not None and predicted_age >= 70:
        signals.append('High predicted body age; prioritize long-term health management')
    elif predicted_age is not None and predicted_age >= 55:
        signals.append('Moderately high predicted body age; lifestyle optimization is recommended')

    if bmi is not None:
        if bmi >= 30:
            signals.append('BMI is in the obesity range')
        elif bmi >= 25:
            signals.append('BMI is elevated')
        elif bmi < 18.5:
            signals.append('BMI is low')

    if systolic is not None and diastolic is not None:
        if systolic >= 140 or diastolic >= 90:
            signals.append('Blood pressure is high')
        elif systolic >= 130 or diastolic >= 80:
            signals.append('Blood pressure is borderline high')

    if cholesterol is not None:
        if cholesterol >= 240:
            signals.append('Cholesterol is high')
        elif cholesterol >= 200:
            signals.append('Cholesterol is borderline high')

    if glucose is not None:
        if glucose >= 126:
            signals.append('Blood glucose is high')
        elif glucose >= 100:
            signals.append('Blood glucose is borderline high')

    activity = str(row.get('Physical Activity Level', '')).lower()
    if activity == 'low':
        signals.append('Physical activity level is low')

    smoking = str(row.get('Smoking Status', '')).lower()
    if smoking == 'current':
        signals.append('Current smoker')
    elif smoking == 'former':
        signals.append('Former smoker; relapse prevention matters')

    alcohol = str(row.get('Alcohol Consumption', '')).lower()
    if alcohol == 'frequent':
        signals.append('Frequent alcohol consumption')

    sleep = str(row.get('Sleep Patterns', '')).lower()
    if sleep == 'insomnia':
        signals.append('Insomnia is reported')
    elif sleep == 'excessive':
        signals.append('Sleep pattern may be excessive')

    if stress is not None and stress >= 7:
        signals.append('Stress level is high')

    if str(row.get('Mental Health Status', '')).lower() in {'poor', 'fair'}:
        signals.append('Mental health self-rating needs attention')

    if not _is_missing(row.get('Chronic Diseases')):
        signals.append(f"Recorded chronic condition: {_clean_value(row.get('Chronic Diseases'))}")

    return signals


def build_lifestyle_profile(row):
    """ LLM  """
    systolic, diastolic = _parse_blood_pressure(row.get('Blood Pressure (s/d)'))
    return {
        'predicted_body_age_years': round(float(row['Predicted Body Age (years)']), 1),
        'gender': _clean_value(row.get('Gender')),
        'bmi': _clean_value(row.get('BMI')),
        'blood_pressure': _clean_value(row.get('Blood Pressure (s/d)')),
        'systolic_bp': systolic,
        'diastolic_bp': diastolic,
        'cholesterol_mg_dl': _clean_value(row.get('Cholesterol Level (mg/dL)')),
        'blood_glucose_mg_dl': _clean_value(row.get('Blood Glucose Level (mg/dL)')),
        'bone_density_g_cm2': _clean_value(row.get('Bone Density (g/cm²)')),
        'physical_activity_level': _clean_value(row.get('Physical Activity Level')),
        'smoking_status': _clean_value(row.get('Smoking Status')),
        'alcohol_consumption': _clean_value(row.get('Alcohol Consumption')),
        'diet': _clean_value(row.get('Diet')),
        'chronic_diseases': _clean_value(row.get('Chronic Diseases')),
        'medication_use': _clean_value(row.get('Medication Use')),
        'family_history': _clean_value(row.get('Family History')),
        'cognitive_function': _clean_value(row.get('Cognitive Function')),
        'mental_health_status': _clean_value(row.get('Mental Health Status')),
        'sleep_patterns': _clean_value(row.get('Sleep Patterns')),
        'stress_levels': _clean_value(row.get('Stress Levels')),
        'pollution_exposure': _clean_value(row.get('Pollution Exposure')),
        'sun_exposure': _clean_value(row.get('Sun Exposure')),
        'key_health_signals': _collect_health_signals(row),
    }



# Final LLM prompt, Qwen2.5-7B-Instruct priority, and local fallback validation.
# English note: see the surrounding code for this project workflow detail.
# English note: see the surrounding code for this project workflow detail.
# Default to the E-drive portable Ollama/Qwen setup used in this project.
# English note: see the surrounding code for this project workflow detail.
os.environ.setdefault('OLLAMA_MODEL', 'qwen2.5:7b-instruct')
os.environ.setdefault('OLLAMA_BASE_URL', 'http://127.0.0.1:11434')
os.environ.setdefault('OLLAMA_HOST', '127.0.0.1:11434')
if os.path.exists(r'E:\CPS5802_Qwen_Ollama\models'):
    os.environ.setdefault('OLLAMA_MODELS', r'E:\CPS5802_Qwen_Ollama\models')

if 'test_data' not in globals():
    test_data = pd.read_csv('Test.csv')

# After the model export section, prefer the best full-train model predictions so recommendations
# are aligned with the strongest validation evidence.
# English note: see the surrounding code for this project workflow detail.
lifestyle_input_df = test_data.copy().reset_index(drop=True)
if 'final_test_predictions_by_model' in globals() and 'best_model_name' in globals() and best_model_name in final_test_predictions_by_model:
    lifestyle_predictions = final_test_predictions_by_model[best_model_name]
    lifestyle_prediction_source = 'full-train {}'.format(best_model_name)
elif 'predictions_df' in globals():
    prediction_column = 'Predicted Age (years)' if 'Predicted Age (years)' in predictions_df.columns else predictions_df.select_dtypes(include=[np.number]).columns[0]
    lifestyle_predictions = predictions_df[prediction_column].values
    lifestyle_prediction_source = 'existing predictions_df'
elif 'y_pred' in globals():
    lifestyle_predictions = np.asarray(y_pred)
    lifestyle_prediction_source = 'existing y_pred'
else:
    # Try saved prediction CSV files before retraining a fallback model.
    # English note: see the surrounding code for this project workflow detail.
    saved_prediction_candidates = [
        ('Linear Regression', Path('model_outputs') / 'LinearRegression' / 'test_predictions.csv'),
        ('Random Forest', Path('model_outputs') / 'RandomForest' / 'test_predictions.csv'),
        ('XGBoost', Path('model_outputs') / 'XGBoost' / 'test_predictions.csv'),
        ('MLP Regressor', Path('model_outputs') / 'MLPRegressor' / 'test_predictions.csv'),
    ]
    lifestyle_predictions = None
    lifestyle_prediction_source = None
    for saved_model_name, saved_prediction_path in saved_prediction_candidates:
        if saved_prediction_path.exists():
            saved_prediction_df = pd.read_csv(str(saved_prediction_path))
            if 'Predicted Body Age (years)' in saved_prediction_df.columns:
                lifestyle_predictions = saved_prediction_df['Predicted Body Age (years)'].values
                lifestyle_prediction_source = 'saved full-train {}'.format(saved_model_name)
                break

    if lifestyle_predictions is None:
        fallback_lifestyle_workflow = AgeModelWorkflow(RandomForestRegressor(n_estimators=120, random_state=RANDOM_STATE, n_jobs=-1), scale_features=True)
        fallback_lifestyle_workflow.fit(X_all, y_all)
        lifestyle_predictions = fallback_lifestyle_workflow.predict(test_data)
        lifestyle_prediction_source = 'fallback RandomForestRegressor'

if len(lifestyle_predictions) != len(lifestyle_input_df):
    raise ValueError('Lifestyle prediction length does not match Test.csv rows.')
lifestyle_input_df['Predicted Body Age (years)'] = np.asarray(lifestyle_predictions).astype(float)


def build_lifestyle_prompt(profile):
    """Professional prompt for educational lifestyle recommendation only.
      """
    profile_json = json.dumps(profile, ensure_ascii=False, indent=2)
    return """
You are a senior preventive medicine and lifestyle medicine physician with 20 years of clinical experience.
You are not making a diagnosis. You are providing educational lifestyle recommendations based only on the synthetic user profile and predicted body age.

Task:
Generate personalized health and lifestyle recommendations using the user's provided data.

Important rules:
1. Do not claim the user has a disease.
2. Do not provide medication instructions.
3. Do not replace a licensed clinician.
4. Clearly state that the dataset is synthetic and the recommendation is for educational purposes only.
5. Use the user's actual data as evidence.
6. Avoid generic advice. Every recommendation should connect to at least one feature, such as BMI, blood pressure, glucose, cholesterol, sleep, stress, activity, smoking, alcohol, or predicted body age.
7. If risk indicators are high, recommend consulting a qualified healthcare professional.

Output format:
1. Brief Profile Summary
2. Main Risk Signals
3. Personalized Recommendations
   - Exercise
   - Diet
   - Sleep and Stress
   - Smoking and Alcohol
   - Follow-up Checks
4. Why These Recommendations Match This User
5. Medical Disclaimer

User profile:
{profile_json}
""".strip().format(profile_json=profile_json)


def _post_json(url, payload, headers=None, timeout=45):
    data = json.dumps(payload).encode('utf-8')
    request_headers = {'Content-Type': 'application/json'}
    if headers:
        request_headers.update(headers)
    req = request.Request(url, data=data, headers=request_headers, method='POST')
    with request.urlopen(req, timeout=timeout) as response:
        return json.loads(response.read().decode('utf-8'))


def _call_openai_llm(prompt, timeout=60):
    api_key = os.getenv('OPENAI_API_KEY')
    if not api_key:
        return None, 'OpenAI skipped: OPENAI_API_KEY is not set'

    base_url = os.getenv('OPENAI_BASE_URL', 'https://api.openai.com/v1').rstrip('/')
    model = os.getenv('OPENAI_MODEL', 'gpt-4o-mini')
    payload = {
        'model': model,
        'messages': [
            {
                'role': 'system',
                'content': 'You provide educational lifestyle recommendations only. Do not diagnose disease, do not use disease labels, and do not give medication instructions.',
            },
            {'role': 'user', 'content': prompt},
        ],
        'temperature': 0.2,
        'max_tokens': 1000,
    }
    try:
        data = _post_json(
            '{}/chat/completions'.format(base_url),
            payload,
            headers={'Authorization': 'Bearer {}'.format(api_key)},
            timeout=timeout,
        )
        content = data['choices'][0]['message']['content'].strip()
        return content, 'OpenAI: {}'.format(model)
    except Exception as exc:
        return None, 'OpenAI call failed: {}'.format(exc)


def _get_ollama_model():
    # English note: see the surrounding code for this project workflow detail.
    # English note: see the surrounding code for this project workflow detail.
    return os.getenv('OLLAMA_MODEL', 'qwen2.5:7b-instruct')


def _sanitize_recommendation_text(text):
    """Replace diagnostic labels and unavailable-age language with safer educational wording.
      """
    replacements = {
        r'indicating an accelerated aging process compared to their actual age': 'indicating a high predicted body-age signal in this synthetic profile',
        r'older-than-actual-age state': 'higher-risk synthetic profile pattern',
        r'older-than-provided synthetic profile state': 'higher-risk synthetic profile pattern',
        r'older-than-[a-z-]*age state': 'higher-risk synthetic profile pattern',
        r'with no provided provided synthetic profile,?': '',
        r'The user, a ([A-Za-z]+)\s+has': r'The user is a \1 and has',
        r'The user, a ([A-Za-z]+) has': r'The user is a \1 and has',
        r'an higher-risk': 'a higher-risk',
        r'advanced biological age': 'high predicted body-age signal',
        r'overweight status': 'an elevated BMI signal',
        r'hyperglycemia': 'an elevated glucose signal',
        r'no provided provided synthetic profile': 'no true age value in Test.csv',
        r'compared to their actual age': 'relative to the model reference pattern',
        r'compared to the user\'s actual age': 'relative to the model reference pattern',
        r'compared to chronological age': 'relative to the model reference pattern',
        r'\bactual-age\b': 'synthetic-profile',
        r'\bactual age\b': 'provided synthetic profile',
        r'\bchronological age\b': 'provided synthetic profile',
        r'\btype 2 diabetes\b': 'an elevated glucose risk signal',
        r'\bobesity\b': 'an elevated BMI signal',
        r'\bhypertension\b': 'an elevated blood-pressure signal',
        r'\bhyperlipidemia\b': 'an elevated cholesterol signal',
        r'\bprediabetes\b': 'an elevated glucose signal',
        r'\bdiabetes\b': 'an elevated glucose signal',
        r'\bosteoporosis\b': 'bone-health risk signals',
        r'\bheart disease\b': 'cardiovascular risk',
        r'\bdiseases\b': 'health conditions',
        r'\bdisease\b': 'health condition',
        r'\bstroke\b': 'cardiovascular/cerebrovascular risk',
        r'\bkidney damage\b': 'kidney-health risk',
        r'\bdiagnosed\b': 'recorded',
        r'immediate attention': 'practical lifestyle attention',
    }
    cleaned = text
    for pattern, replacement in replacements.items():
        cleaned = re.sub(pattern, replacement, cleaned, flags=re.IGNORECASE)
    cleaned = re.sub(r'\s+,', ',', cleaned)
    cleaned = re.sub(r' {2,}', ' ', cleaned)
    return cleaned


# English note: see the surrounding code for this project workflow detail.
# English note: see the surrounding code for this project workflow detail.
def _call_ollama_llm(prompt, timeout=420):
    model = _get_ollama_model()
    base_url = os.getenv('OLLAMA_BASE_URL', 'http://127.0.0.1:11434').rstrip('/')
    payload = {
        'model': model,
        'messages': [
            {
                'role': 'system',
                'content': 'You provide educational lifestyle recommendations only. You do not diagnose disease or give medication instructions.',
            },
            {'role': 'user', 'content': prompt},
        ],
        'stream': False,
        'options': {'temperature': 0.2, 'num_predict': 700},
    }
    try:
        data = _post_json('{}/api/chat'.format(base_url), payload, timeout=timeout)
        content = _sanitize_recommendation_text(data['message']['content'].strip())
        return content, 'Ollama Qwen2.5: {}'.format(model)
    except Exception as exc:
        return None, 'Ollama Qwen2.5 call failed for model {}: {}'.format(model, exc)


def _rule_based_recommendation(profile):
    signals = profile.get('key_health_signals', [])
    bmi = _clean_value(profile.get('bmi'))
    blood_pressure = _clean_value(profile.get('blood_pressure'))
    cholesterol = _clean_value(profile.get('cholesterol_mg_dl'))
    glucose = _clean_value(profile.get('blood_glucose_mg_dl'))
    activity = _clean_value(profile.get('physical_activity_level'))
    smoking = _clean_value(profile.get('smoking_status'))
    alcohol = _clean_value(profile.get('alcohol_consumption'))
    sleep = _clean_value(profile.get('sleep_patterns'))
    stress = _clean_value(profile.get('stress_levels'))
    predicted_age = profile.get('predicted_body_age_years')

    exercise_advice = 'Match exercise to the recorded activity level ({}): aim for regular aerobic activity and add two weekly strength sessions if tolerated.'.format(activity)
    if str(activity).lower() == 'low':
        exercise_advice = 'Because physical activity is low, start with 20-30 minutes of brisk walking or cycling on most days and add light strength training twice weekly.'

    diet_advice = 'Use BMI ({}), cholesterol ({} mg/dL), and glucose ({} mg/dL) as diet signals: emphasize vegetables, legumes, whole grains, lean protein, and limit sugary drinks, fried foods, and refined carbohydrates.'.format(bmi, cholesterol, glucose)
    sleep_stress_advice = 'Sleep pattern is {} and stress level is {}; keep a consistent wake time, reduce late-night screens, and use breathing, stretching, or mindfulness routines.'.format(sleep, stress)
    smoking_alcohol_advice = 'Smoking status is {} and alcohol use is {}; avoid smoking exposure, prevent relapse if former smoker, and keep alcohol low or occasional.'.format(smoking, alcohol)
    follow_up_advice = 'Because blood pressure is {}, glucose is {}, cholesterol is {}, and predicted body age is {}, consider periodic checks and consult a qualified healthcare professional if readings remain high or symptoms occur.'.format(blood_pressure, glucose, cholesterol, predicted_age)

    if not signals:
        signals = ['No major rule-based signal was detected, so recommendations focus on maintaining healthy routines.']

    lines = [
        '1. Brief Profile Summary',
        '- Synthetic profile with predicted body age of {} years. Key values include BMI {}, blood pressure {}, glucose {} mg/dL, cholesterol {} mg/dL, activity {}, sleep {}, and stress {}.'.format(predicted_age, bmi, blood_pressure, glucose, cholesterol, activity, sleep, stress),
        '',
        '2. Main Risk Signals',
    ]
    lines.extend(['- {}'.format(signal) for signal in signals[:8]])
    lines.extend([
        '',
        '3. Personalized Recommendations',
        '- Exercise: {}'.format(exercise_advice),
        '- Diet: {}'.format(diet_advice),
        '- Sleep and Stress: {}'.format(sleep_stress_advice),
        '- Smoking and Alcohol: {}'.format(smoking_alcohol_advice),
        '- Follow-up Checks: {}'.format(follow_up_advice),
        '',
        '4. Why These Recommendations Match This User',
        '- The advice is linked to the provided synthetic features: predicted body age, BMI, blood pressure, glucose, cholesterol, activity, smoking, alcohol, sleep, and stress. Higher-risk signals receive more conservative follow-up guidance.',
        '',
        '5. Medical Disclaimer',
        '- This dataset is synthetic and this recommendation is for educational purposes only. It is not a diagnosis, does not provide medication instructions, and does not replace a licensed clinician.',
    ])
    return '\n'.join(lines)


def generate_lifestyle_recommendation(person_index=0, use_llm=True):
    if person_index < 0 or person_index >= len(lifestyle_input_df):
        raise IndexError('person_index must be between 0 and {}'.format(len(lifestyle_input_df) - 1))

    row = lifestyle_input_df.iloc[person_index]
    profile = build_lifestyle_profile(row)
    prompt = build_lifestyle_prompt(profile)
    llm_logs = []

    if use_llm:
        if os.getenv('OPENAI_API_KEY'):
            content, provider_message = _call_openai_llm(prompt)
            llm_logs.append(provider_message)
            if content:
                return {
                    'person_index': person_index,
                    'predicted_body_age': profile['predicted_body_age_years'],
                    'provider': provider_message,
                    'Recommendation Source': provider_message,
                    'recommendation': content,
                    'llm_logs': llm_logs,
                }
        else:
            llm_logs.append('OpenAI skipped: OPENAI_API_KEY is not set')

        content, provider_message = _call_ollama_llm(prompt)
        llm_logs.append(provider_message)
        if content:
            return {
                'person_index': person_index,
                'predicted_body_age': profile['predicted_body_age_years'],
                'provider': provider_message,
                'Recommendation Source': provider_message,
                'recommendation': content,
                'llm_logs': llm_logs,
            }

    return {
        'person_index': person_index,
        'predicted_body_age': profile['predicted_body_age_years'],
        'provider': 'Local rule-based fallback',
        'Recommendation Source': 'Local rule-based fallback',
        'recommendation': _rule_based_recommendation(profile),
        'llm_logs': llm_logs,
    }


print('Final recommendation workflow is ready.')
print('Lifestyle prediction source: {}'.format(lifestyle_prediction_source))
print('Default Ollama model: {}'.format(_get_ollama_model()))
````

The examples come from `student_sample_outputs.csv`, which was produced by running the student model on held-out test prompts and saving its generated responses side by side with the Qwen2.5-7B teacher responses.

The saved examples are used here so the notebook remains lightweight and reproducible. A live local inference command is included for verification, but the main notebook display uses the saved evaluation outputs.

````python
# Step 12: display several LoRA student recommendation examples safely.
# The saved responses contain Markdown syntax such as **bold headings**. To avoid broken
# formatting when long text is truncated, render model responses as escaped HTML text blocks.
from html import escape
from IPython.display import HTML

sample_outputs = pd.read_csv(PROJECT_LORA_OUTPUT_DIR / "student_sample_outputs.csv")

example_count = min(3, len(sample_outputs))
example_rows = sample_outputs.head(example_count).copy()
example_overview = example_rows[["distillation_id", "person_index", "predicted_body_age_years"]]
display(example_overview)


def clean_response_text(text):
    text = str(text).replace("\r\n", "\n").replace("\r", "\n")
    lines = [line.rstrip() for line in text.splitlines()]
    return "\n".join(lines).strip()


def make_safe_preview(text, max_chars=3500):
    text = clean_response_text(text)
    if len(text) <= max_chars:
        return text
    cut_at = text.rfind("\n", 0, max_chars)
    if cut_at < int(max_chars * 0.65):
        cut_at = max_chars
    return (
        text[:cut_at].rstrip()
        + "\n\n[Display truncated. See model_outputs/lora_student/student_sample_outputs.csv for the full saved response.]"
    )


def display_response_block(title, text, max_chars=3500):
    safe_title = escape(title)
    safe_text = escape(make_safe_preview(text, max_chars=max_chars))
    html = f"""
    <div style="margin: 10px 0 22px 0;">
      <div style="font-weight: 700; margin-bottom: 6px;">{safe_title}</div>
      <pre style="white-space: pre-wrap; font-family: Consolas, 'Courier New', monospace; line-height: 1.45;
                  background: #111827; color: #f3f4f6; border: 1px solid #374151;
                  border-radius: 6px; padding: 12px; max-height: 520px; overflow: auto;">{safe_text}</pre>
    </div>
    """
    display(HTML(html))


for _, row in example_rows.iterrows():
    display(Markdown(f"### LoRA Student Example: distillation_id={int(row['distillation_id'])}, person_index={int(row['person_index'])}"))
    display(Markdown(f"**Predicted body age:** {row['predicted_body_age_years']:.1f} years"))
    display_response_block("Qwen2.5-0.5B LoRA student saved output", row["student_response"])
    display_response_block("Qwen2.5-7B teacher saved output", row["teacher_response"])

live_inference_command = "\n".join([
    "$env:QWEN_LORA_ROOT='E:\\CPS5802_Qwen_LoRA'",
    "E:\\CPS5802_Qwen_LoRA\\.venv_qwen_lora\\Scripts\\python.exe run_student_inference.py --sample-index 0 --max-new-tokens 700 --dtype fp16",
])

display(Markdown("**Optional live inference command:**\n```powershell\n" + live_inference_command + "\n```"))
````

## Step 13: Qwen2.5-7B Teacher vs Qwen2.5-0.5B LoRA Student Comparison

This step compares the larger teacher model and the smaller LoRA student on the same held-out evaluation samples. The goal is not to claim clinical accuracy. The goal is to measure how closely the lightweight local student preserves the teacher's recommendation quality under the project's structured rubric.

Current headline result:

- Qwen2.5-0.5B LoRA student average quality score: **91.3585**
- Qwen2.5-7B teacher average quality score: **92.2295**
- Student quality retention relative to teacher: about **99.06%**

This means the student model closely matches the teacher on recommendation structure, risk-signal coverage, actionability, and profile-grounded explanation while being much smaller and easier to run locally.

````python
# Step 13: compute teacher-student summary, retention, and per-sample deltas.
summary_df = pd.read_csv(PROJECT_LORA_OUTPUT_DIR / "student_eval_summary.csv")
scores_df = pd.read_csv(PROJECT_LORA_OUTPUT_DIR / "student_eval_scores.csv")

student_name = "Qwen2.5-0.5B LoRA student"
teacher_name = "Qwen2.5-7B teacher"

summary_ordered = summary_df.set_index("response_source").loc[[student_name, teacher_name]].reset_index()
display(summary_ordered)

student_quality = float(summary_ordered.loc[summary_ordered["response_source"] == student_name, "quality_score"].iloc[0])
teacher_quality = float(summary_ordered.loc[summary_ordered["response_source"] == teacher_name, "quality_score"].iloc[0])
quality_gap = teacher_quality - student_quality
retention = student_quality / teacher_quality * 100

headline_df = pd.DataFrame([
    {"Metric": "Student average quality score", "Value": student_quality},
    {"Metric": "Teacher average quality score", "Value": teacher_quality},
    {"Metric": "Teacher - student score gap", "Value": quality_gap},
    {"Metric": "Student retention of teacher score (%)", "Value": retention},
])
display(headline_df)

pivot_quality = scores_df.pivot(index="distillation_id", columns="response_source", values="quality_score").reset_index()
pivot_quality["student_minus_teacher"] = pivot_quality[student_name] - pivot_quality[teacher_name]

delta_summary = pivot_quality["student_minus_teacher"].describe().to_frame("student_minus_teacher_quality_delta")
display(delta_summary)

fig, axes = plt.subplots(1, 2, figsize=(13, 4.8))
axes[0].bar(["0.5B LoRA student", "7B teacher"], [student_quality, teacher_quality], color=["#4C78A8", "#F28E2B"])
axes[0].set_title("Average Recommendation Quality Score")
axes[0].set_ylim(0, 100)
axes[0].set_ylabel("Rubric score")
for i, value in enumerate([student_quality, teacher_quality]):
    axes[0].text(i, value + 1, f"{value:.2f}", ha="center")

axes[1].hist(pivot_quality["student_minus_teacher"], bins=18, color="#59A14F", edgecolor="white")
axes[1].axvline(0, color="#333333", linewidth=1)
axes[1].set_title("Per-Sample Quality Delta")
axes[1].set_xlabel("Student - teacher quality score")
axes[1].set_ylabel("Sample count")
fig.tight_layout()
plt.show()
````

````python
# Step 15 setup: reusable loaders, paths, plotting helpers, and text-metric helpers.
from pathlib import Path
import json
import math
import re
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, Image, HTML

warnings.filterwarnings("ignore", category=UserWarning)
plt.rcParams.update({
    "figure.dpi": 120,
    "axes.grid": True,
    "grid.alpha": 0.25,
    "axes.spines.top": False,
    "axes.spines.right": False,
})

PROJECT_ROOT = Path.cwd()
OUTPUT_DIR = PROJECT_ROOT / "model_outputs"
DISTILLATION_DIR = OUTPUT_DIR / "distillation"
FIGURE_DIR = OUTPUT_DIR / "figures"
REPORT_DIR = OUTPUT_DIR / "reports"
STUDENT_FILTERED_DIR = OUTPUT_DIR / "student_filtered_passed_inference"
FINAL_JUDGE_DIR = PROJECT_ROOT / "final judge data"
CHECKLIST_DIR = PROJECT_ROOT / "checklist" / "Training data filtering checklist"
TEACHER_STUDENT_OUTPUT_DIR = next((p for p in PROJECT_ROOT.iterdir() if p.is_dir() and p.name.startswith("teacher")), PROJECT_ROOT / "teacher stu_output")

FIGURE_DIR.mkdir(parents=True, exist_ok=True)
REPORT_DIR.mkdir(parents=True, exist_ok=True)

QWEN_TEACHER_CSV = DISTILLATION_DIR / "qwen25_7b_distillation_dataset_full.csv"
LLAMA_TEACHER_CSV = DISTILLATION_DIR / "llama31_8b_distillation_dataset_full.csv"
MULTITEACHER_CSV = DISTILLATION_DIR / "multiteacher_qwen25_7b_llama31_8b_dataset_full.csv"

LORA_QWEN_STUDENT_DIR = OUTPUT_DIR / "lora_student"
LORA_LLAMA_STUDENT_DIR = OUTPUT_DIR / "lora_student_llama31_teacher"
LORA_MULTI_STUDENT_DIR = OUTPUT_DIR / "lora_student_multiteacher_qwen_llama31"

STUDENT_OUTPUT_PATHS = {
    "student_qwen_from_qwen": STUDENT_FILTERED_DIR / "student_from_qwen" / "student_sample_outputs.csv",
    "student_qwen_from_llama": STUDENT_FILTERED_DIR / "student_from_llama" / "student_sample_outputs.csv",
    "student_qwen_from_both": STUDENT_FILTERED_DIR / "student_from_both" / "student_sample_outputs.csv",
}
STUDENT_EVAL_SUMMARY_PATHS = {
    "student_qwen_from_qwen": STUDENT_FILTERED_DIR / "student_from_qwen" / "student_eval_summary.csv",
    "student_qwen_from_llama": STUDENT_FILTERED_DIR / "student_from_llama" / "student_eval_summary.csv",
    "student_qwen_from_both": STUDENT_FILTERED_DIR / "student_from_both" / "student_eval_summary.csv",
}

MODEL_LABELS = {
    "teacher_qwen25_7b": "Teacher: Qwen2.5-7B",
    "teacher_llama31_8b": "Teacher: Llama 3.1 8B",
    "student_qwen_from_qwen": "Student: 0.5B from Qwen",
    "student_qwen_from_llama": "Student: 0.5B from Llama",
    "student_qwen_from_both": "Student: 0.5B from Both",
}
MODEL_ORDER = list(MODEL_LABELS.keys())
MODEL_COLORS = {
    "teacher_qwen25_7b": "#4C78A8",
    "teacher_llama31_8b": "#F58518",
    "student_qwen_from_qwen": "#54A24B",
    "student_qwen_from_llama": "#B279A2",
    "student_qwen_from_both": "#E45756",
}

JUDGE_DIRS = {
    "teacher_qwen25_7b": FINAL_JUDGE_DIR / "01_teacher_Qwen2.5-7B",
    "teacher_llama31_8b": FINAL_JUDGE_DIR / "02_teacher_Llama_3.1_8B",
    "student_qwen_from_qwen": FINAL_JUDGE_DIR / "03_student_Qwen2.5-0.5B_from_Qwen2.5-7B",
    "student_qwen_from_llama": FINAL_JUDGE_DIR / "04_student_Qwen2.5-0.5B_from_Llama_3.1_8B",
    "student_qwen_from_both": FINAL_JUDGE_DIR / "05_student_Qwen2.5-0.5B_from_Qwen2.5-7B_plus_Llama_3.1_8B",
}

REQUIRED_HEADINGS = [
    "Brief Profile Summary",
    "Main Risk Signals",
    "Personalized Recommendations",
    "Why These Recommendations Match This User",
    "Medical Disclaimer",
]
ACTION_TERMS = [
    "increase", "reduce", "monitor", "consult", "schedule", "limit", "avoid", "practice", "track", "prioritize",
    "maintain", "replace", "review", "check", "aim", "consider", "follow", "choose", "build", "discuss",
]
EXPLANATION_TERMS = [
    "because", "therefore", "due to", "given", "since", "which may", "this suggests", "this helps", "as a result", "matches",
]
MEDICAL_DISCLAIMER_TERMS = [
    "medical disclaimer", "healthcare professional", "doctor", "clinician", "not a diagnosis", "professional medical advice", "consult",
]


def read_csv_safe(path, **kwargs):
    path = Path(path)
    if not path.exists():
        raise FileNotFoundError(f"Missing expected CSV: {path}")
    with path.open("r", encoding="utf-8-sig", newline="") as handle:
        return pd.read_csv(handle, **kwargs)


def read_json_safe(path):
    path = Path(path)
    if not path.exists():
        return {}
    with path.open("r", encoding="utf-8-sig") as handle:
        return json.load(handle)


def count_jsonl(path):
    path = Path(path)
    if not path.exists():
        return 0
    with path.open("r", encoding="utf-8-sig") as handle:
        return sum(1 for _ in handle)


def save_figure(filename):
    path = FIGURE_DIR / filename
    plt.tight_layout()
    plt.savefig(path, dpi=180, bbox_inches="tight")
    plt.show()
    display(Markdown(f"Saved figure: `{path.relative_to(PROJECT_ROOT)}`"))
    return path


def as_numeric(series):
    return pd.to_numeric(series, errors="coerce")


def status_counts(df):
    return df["quality_status"].fillna("missing").value_counts().reindex(["passed", "review_needed", "missing"], fill_value=0)


def clean_text(text):
    return str(text if pd.notna(text) else "").replace("\r\n", "\n").replace("\r", "\n").strip()


def response_text_metrics(text):
    text = clean_text(text)
    lower = text.lower()
    words = re.findall(r"[A-Za-z]+(?:'[A-Za-z]+)?|\d+(?:\.\d+)?", text)
    heading_count = sum(heading.lower() in lower for heading in REQUIRED_HEADINGS)
    bullet_count = sum(1 for line in text.splitlines() if re.match(r"^\s*(?:[-*•]|\d+[.)])\s+", line))
    action_count = sum(len(re.findall(r"\b" + re.escape(term) + r"\b", lower)) for term in ACTION_TERMS)
    explanation_count = sum(lower.count(term) for term in EXPLANATION_TERMS)
    numeric_evidence_count = len(re.findall(r"\b\d+(?:\.\d+)?\b", text))
    medical_disclaimer_present = int(any(term in lower for term in MEDICAL_DISCLAIMER_TERMS))
    return {
        "word_count": len(words),
        "char_count": len(text),
        "line_count": len([line for line in text.splitlines() if line.strip()]),
        "bullet_count": bullet_count,
        "required_heading_count": heading_count,
        "required_heading_coverage": heading_count / len(REQUIRED_HEADINGS),
        "action_verb_count": action_count,
        "explanation_connector_count": explanation_count,
        "numeric_evidence_count": numeric_evidence_count,
        "medical_disclaimer_present": medical_disclaimer_present,
    }


def metrics_for_text_series(series):
    return pd.DataFrame(series.fillna("").map(response_text_metrics).tolist(), index=series.index)


def make_text_metric_frame(df, text_col, model_key, model_label):
    metric_df = metrics_for_text_series(df[text_col])
    metric_df.insert(0, "person_index", df["person_index"].values)
    metric_df.insert(0, "model_label", model_label)
    metric_df.insert(0, "model_key", model_key)
    return metric_df


def plot_bar_labels(ax, fmt="{:.1f}", pad=0.5, rotation=0):
    for patch in ax.patches:
        height = patch.get_height()
        if np.isfinite(height):
            ax.text(
                patch.get_x() + patch.get_width() / 2,
                height + pad,
                fmt.format(height),
                ha="center",
                va="bottom",
                fontsize=8,
                rotation=rotation,
            )


print("Step 15+ analysis setup ready.")
print("Project root:", PROJECT_ROOT)
print("Teacher/student organized output dir:", TEACHER_STUDENT_OUTPUT_DIR)
````

````python
# Step 17: verify and visualize the multi-teacher distillation dataset.
multiteacher_df = read_csv_safe(MULTITEACHER_CSV)
multiteacher_metadata = read_json_safe(LORA_MULTI_STUDENT_DIR / "data" / "dataset_metadata.json")

qwen_passed_ids = set(qwen_teacher_df.loc[qwen_teacher_df["quality_status"] == "passed", "person_index"].astype(int))
llama_passed_ids = set(llama_teacher_df.loc[llama_teacher_df["quality_status"] == "passed", "person_index"].astype(int))
all_ids = set(pd.concat([qwen_teacher_df["person_index"], llama_teacher_df["person_index"]]).astype(int))
common_passed_ids = qwen_passed_ids & llama_passed_ids
qwen_only_ids = qwen_passed_ids - llama_passed_ids
llama_only_ids = llama_passed_ids - qwen_passed_ids
neither_passed_ids = all_ids - (qwen_passed_ids | llama_passed_ids)

overlap_df = pd.DataFrame([
    {"subset": "Both teachers passed", "rows": len(common_passed_ids)},
    {"subset": "Qwen only passed", "rows": len(qwen_only_ids)},
    {"subset": "Llama only passed", "rows": len(llama_only_ids)},
    {"subset": "Neither passed", "rows": len(neither_passed_ids)},
])
source_counts = multiteacher_df["source_teacher_family"].value_counts().rename_axis("source_teacher_family").reset_index(name="rows")

multi_split_rows = []
for split_name, split_info in multiteacher_metadata.get("splits", {}).items():
    multi_split_rows.append({"split": split_name, "rows": split_info.get("rows"), "approx_total_tokens_chars_4": split_info.get("approx_total_tokens_chars_4")})
multi_split_df = pd.DataFrame(multi_split_rows)

summary_multi_df = pd.DataFrame([
    {"metric": "Multi-teacher rows", "value": len(multiteacher_df)},
    {"metric": "Unique person_index values", "value": multiteacher_df["person_index"].nunique()},
    {"metric": "Qwen passed rows", "value": len(qwen_passed_ids)},
    {"metric": "Llama passed rows", "value": len(llama_passed_ids)},
    {"metric": "Common-passed profile count", "value": len(common_passed_ids)},
])

display(summary_multi_df)
display(source_counts)
display(overlap_df)
display(multi_split_df)

fig, axes = plt.subplots(2, 2, figsize=(14, 9))
axes[0, 0].bar(source_counts["source_teacher_family"], source_counts["rows"], color=["#4C78A8", "#F58518"])
axes[0, 0].set_title("Rows by Teacher Source in Multi-Teacher Dataset")
axes[0, 0].set_ylabel("Rows")
plot_bar_labels(axes[0, 0], fmt="{:.0f}", pad=40)

axes[0, 1].bar(overlap_df["subset"], overlap_df["rows"], color=["#54A24B", "#4C78A8", "#F58518", "#E45756"])
axes[0, 1].set_title("Teacher Passed-Row Overlap")
axes[0, 1].set_ylabel("Profiles")
axes[0, 1].tick_params(axis="x", rotation=25)
plot_bar_labels(axes[0, 1], fmt="{:.0f}", pad=35)

for source_name, group in multiteacher_df.groupby("source_teacher_family"):
    axes[1, 0].hist(as_numeric(group["predicted_body_age_years"]).dropna(), bins=30, alpha=0.65, label=source_name)
axes[1, 0].set_title("Predicted Body-Age Coverage by Teacher Source")
axes[1, 0].set_xlabel("Predicted body age (years)")
axes[1, 0].set_ylabel("Rows")
axes[1, 0].legend()

axes[1, 1].bar(multi_split_df["split"], multi_split_df["rows"], color=["#4C78A8", "#F58518", "#54A24B"])
axes[1, 1].set_title("Multi-Teacher Student Training Split")
axes[1, 1].set_ylabel("Rows")
plot_bar_labels(axes[1, 1], fmt="{:.0f}", pad=60)

save_figure("step17_multiteacher_dataset_construction.png")
````

A third student model was trained from the combined Qwen + Llama teacher dataset. This gives three student variants for comparison:

- Student distilled from Llama 3.1 8B.

````python
# Step 18: compare the three student variants using project-rubric evaluation summaries.
def load_student_eval_summary_table():
    rows = []
    for model_key, path in STUDENT_EVAL_SUMMARY_PATHS.items():
        if not Path(path).exists():
            continue
        df = read_csv_safe(path).copy()
        df["model_key"] = model_key
        df["model_label_short"] = MODEL_LABELS[model_key]
        rows.append(df)
    return pd.concat(rows, ignore_index=True) if rows else pd.DataFrame()

student_eval_all = load_student_eval_summary_table()
student_rows = student_eval_all[student_eval_all["response_source"].str.contains("student", case=False, na=False)].copy()
reference_rows = student_eval_all[~student_eval_all["response_source"].str.contains("student", case=False, na=False)].copy()

metric_cols = [
    "quality_score", "word_count", "bullet_count", "numeric_evidence_count", "action_verb_count",
    "explanation_connector_count", "feature_evidence_count", "key_signal_coverage", "forbidden_term_count",
    "medical_disclaimer_present",
]
metric_cols = [col for col in metric_cols if col in student_rows.columns]

student_project_rubric_summary = student_rows[["model_key", "model_label_short", "response_source"] + metric_cols].reset_index(drop=True)
display(student_project_rubric_summary)
student_project_rubric_summary.to_csv(REPORT_DIR / "step18_three_student_project_rubric_summary.csv", index=False, encoding="utf-8-sig")

fig, axes = plt.subplots(2, 2, figsize=(15, 9))
quality_plot = student_project_rubric_summary.sort_values("quality_score", ascending=False)
axes[0, 0].bar(quality_plot["model_label_short"], quality_plot["quality_score"],
               color=[MODEL_COLORS.get(k, "#999999") for k in quality_plot["model_key"]])
axes[0, 0].set_title("Three Student Models: Project Rubric Quality")
axes[0, 0].set_ylim(0, 100)
axes[0, 0].set_ylabel("Quality score")
axes[0, 0].tick_params(axis="x", rotation=20)
plot_bar_labels(axes[0, 0], fmt="{:.2f}", pad=1)

selected_heatmap_metrics = ["quality_score", "word_count", "bullet_count", "action_verb_count", "explanation_connector_count", "feature_evidence_count", "key_signal_coverage"]
selected_heatmap_metrics = [m for m in selected_heatmap_metrics if m in student_project_rubric_summary.columns]
heat_raw = student_project_rubric_summary.set_index("model_label_short")[selected_heatmap_metrics]
heat_norm = heat_raw.copy().astype(float)
for col in heat_norm.columns:
    col_min = heat_norm[col].min()
    col_max = heat_norm[col].max()
    heat_norm[col] = 0.5 if math.isclose(col_min, col_max) else (heat_norm[col] - col_min) / (col_max - col_min)
im = axes[0, 1].imshow(heat_norm, aspect="auto", cmap="viridis")
axes[0, 1].set_title("Normalized Metric Heatmap Across Students")
axes[0, 1].set_xticks(range(len(heat_norm.columns)))
axes[0, 1].set_xticklabels(heat_norm.columns, rotation=35, ha="right")
axes[0, 1].set_yticks(range(len(heat_norm.index)))
axes[0, 1].set_yticklabels(heat_norm.index)
fig.colorbar(im, ax=axes[0, 1], fraction=0.046, pad=0.04)

if not reference_rows.empty:
    reference = reference_rows.groupby("model_key")[metric_cols].mean()
    student_metric_indexed = student_project_rubric_summary.set_index("model_key")[metric_cols]
    delta_rows = []
    for key in student_metric_indexed.index:
        if key in reference.index:
            delta = student_metric_indexed.loc[key] - reference.loc[key]
            delta["model_label_short"] = MODEL_LABELS[key]
            delta_rows.append(delta)
    delta_df = pd.DataFrame(delta_rows).set_index("model_label_short") if delta_rows else pd.DataFrame()
    if "quality_score" in delta_df.columns:
        axes[1, 0].bar(delta_df.index, delta_df["quality_score"], color="#E45756")
        axes[1, 0].axhline(0, color="#333333", linewidth=1)
        axes[1, 0].set_title("Quality Delta vs Prompt Source Reference")
        axes[1, 0].set_ylabel("Student - reference")
        axes[1, 0].tick_params(axis="x", rotation=20)
else:
    axes[1, 0].axis("off")

if {"word_count", "quality_score"}.issubset(student_project_rubric_summary.columns):
    axes[1, 1].scatter(student_project_rubric_summary["word_count"], student_project_rubric_summary["quality_score"],
                       s=160, color=[MODEL_COLORS[k] for k in student_project_rubric_summary["model_key"]])
    for _, row in student_project_rubric_summary.iterrows():
        axes[1, 1].annotate(row["model_label_short"], (row["word_count"], row["quality_score"]), xytext=(6, 4), textcoords="offset points", fontsize=8)
    axes[1, 1].set_title("Detail Level vs Project Rubric Quality")
    axes[1, 1].set_xlabel("Average word count")
    axes[1, 1].set_ylabel("Quality score")

save_figure("step18_three_student_project_rubric_comparison.png")
````

## Step 21: Five-Model Output Organization

The final project output contains five comparable response sources:

1. Teacher: Qwen2.5-7B.
2. Teacher: Llama 3.1 8B.
3. Student: Qwen2.5-0.5B from Qwen2.5-7B.
4. Student: Qwen2.5-0.5B from Llama 3.1 8B.
5. Student: Qwen2.5-0.5B from Qwen2.5-7B + Llama 3.1 8B.

This step builds a unified long-format table so response length, structure, actionability, and safety wording can be compared consistently before using the MedGemma judge.


````python
# Step 21: build a unified five-model output table and compare response text behavior.
def build_five_model_output_table():
    common_ids = set(common_passed_df["person_index"].astype(int))
    frames = []

    qwen = qwen_teacher_df[qwen_teacher_df["person_index"].astype(int).isin(common_ids)].copy()
    qwen = qwen[["person_index", "distillation_id", "predicted_body_age_years", "teacher_response"]].rename(columns={"teacher_response": "response_text"})
    qwen["model_key"] = "teacher_qwen25_7b"
    frames.append(qwen)

    llama = llama_teacher_df[llama_teacher_df["person_index"].astype(int).isin(common_ids)].copy()
    llama = llama[["person_index", "distillation_id", "predicted_body_age_years", "teacher_response"]].rename(columns={"teacher_response": "response_text"})
    llama["model_key"] = "teacher_llama31_8b"
    frames.append(llama)

    for model_key, path in STUDENT_OUTPUT_PATHS.items():
        df = read_csv_safe(path).copy()
        df = df[["person_index", "distillation_id", "predicted_body_age_years", "student_response"]].rename(columns={"student_response": "response_text"})
        df["model_key"] = model_key
        frames.append(df)

    out = pd.concat(frames, ignore_index=True)
    out["model_label_short"] = out["model_key"].map(MODEL_LABELS)
    out["model_order"] = out["model_key"].map({key: i for i, key in enumerate(MODEL_ORDER)})
    return out.sort_values(["model_order", "person_index"]).reset_index(drop=True)

five_model_outputs_df = build_five_model_output_table()
five_model_outputs_df.to_csv(REPORT_DIR / "step21_five_model_common_passed_outputs_long.csv", index=False, encoding="utf-8-sig")

five_model_text_metrics_df = []
for model_key, group in five_model_outputs_df.groupby("model_key"):
    five_model_text_metrics_df.append(make_text_metric_frame(group, "response_text", model_key, MODEL_LABELS[model_key]))
five_model_text_metrics_df = pd.concat(five_model_text_metrics_df, ignore_index=True)
five_model_text_metrics_df.to_csv(REPORT_DIR / "step21_five_model_text_metrics.csv", index=False, encoding="utf-8-sig")

five_model_row_check = (
    five_model_outputs_df
    .groupby(["model_key", "model_label_short"])
    .apply(lambda group: pd.Series({
        "rows": len(group),
        "unique_person_index": group["person_index"].nunique(),
        "empty_response": int(group["response_text"].fillna("").astype(str).str.strip().eq("").sum()),
    }))
    .reset_index()
)
display(five_model_row_check)

five_metric_summary = five_model_text_metrics_df.groupby(["model_key", "model_label"])[[
    "word_count", "bullet_count", "required_heading_coverage", "action_verb_count",
    "explanation_connector_count", "numeric_evidence_count", "medical_disclaimer_present",
]].mean().reset_index()
five_metric_summary["model_order"] = five_metric_summary["model_key"].map({key: i for i, key in enumerate(MODEL_ORDER)})
five_metric_summary = five_metric_summary.sort_values("model_order")
display(five_metric_summary.drop(columns="model_order"))

fig, axes = plt.subplots(2, 3, figsize=(18, 9.5))
ordered_labels = [MODEL_LABELS[k] for k in MODEL_ORDER]

axes[0, 0].bar(ordered_labels, five_model_row_check.set_index("model_key").loc[MODEL_ORDER, "rows"], color=[MODEL_COLORS[k] for k in MODEL_ORDER])
axes[0, 0].set_title("Rows in Unified Five-Model Table")
axes[0, 0].set_ylabel("Rows")
axes[0, 0].tick_params(axis="x", rotation=25)
plot_bar_labels(axes[0, 0], fmt="{:.0f}", pad=35)

word_data = [five_model_text_metrics_df.loc[five_model_text_metrics_df["model_key"] == k, "word_count"] for k in MODEL_ORDER]
axes[0, 1].boxplot(word_data, labels=ordered_labels, patch_artist=True)
axes[0, 1].set_title("Response Word Count by Model")
axes[0, 1].set_ylabel("Words")
axes[0, 1].tick_params(axis="x", rotation=25)

heading_rate = five_metric_summary.set_index("model_key").loc[MODEL_ORDER, "required_heading_coverage"] * 100
axes[0, 2].bar(ordered_labels, heading_rate, color=[MODEL_COLORS[k] for k in MODEL_ORDER])
axes[0, 2].set_title("Required Heading Coverage by Model")
axes[0, 2].set_ylim(0, 105)
axes[0, 2].set_ylabel("Coverage (%)")
axes[0, 2].tick_params(axis="x", rotation=25)
plot_bar_labels(axes[0, 2], fmt="{:.1f}", pad=1)

five_metric_summary.set_index("model_label")[["action_verb_count", "explanation_connector_count"]].plot(kind="bar", ax=axes[1, 0], color=["#4C78A8", "#F58518"])
axes[1, 0].set_title("Actionability and Explanation Detail")
axes[1, 0].set_ylabel("Average count")
axes[1, 0].tick_params(axis="x", rotation=25)

five_metric_summary.set_index("model_label")[["numeric_evidence_count", "bullet_count"]].plot(kind="bar", ax=axes[1, 1], color=["#72B7B2", "#B279A2"])
axes[1, 1].set_title("Numeric Evidence and Bullet Structure")
axes[1, 1].set_ylabel("Average count")
axes[1, 1].tick_params(axis="x", rotation=25)

disc_rate = five_metric_summary.set_index("model_key").loc[MODEL_ORDER, "medical_disclaimer_present"] * 100
axes[1, 2].bar(ordered_labels, disc_rate, color=[MODEL_COLORS[k] for k in MODEL_ORDER])
axes[1, 2].set_title("Medical Disclaimer Mention Rate")
axes[1, 2].set_ylim(0, 105)
axes[1, 2].set_ylabel("Responses (%)")
axes[1, 2].tick_params(axis="x", rotation=25)
plot_bar_labels(axes[1, 2], fmt="{:.1f}", pad=1)

save_figure("step21_five_model_output_text_metric_comparison.png")
````

## Step 22: MedGemma 27B Q4 Judge Evaluation

The previous project-rubric scores are useful because they are deterministic and transparent. However, the project also adds an independent LLM judge: **MedGemma 27B Q4 judge**.

This step compares all five response sources on the same 2933 common-passed profiles. The judge score should still be interpreted as recommendation-quality evaluation, not clinical validation.


````python
# Step 22: load MedGemma judge summaries and create five-model comparison views.
def load_judge_summary_all():
    frames = []
    for model_key, judge_dir in JUDGE_DIRS.items():
        path = judge_dir / "medgemma_judge_summary.csv"
        df = read_csv_safe(path).copy()
        df["model_key"] = model_key
        df["model_label_short"] = MODEL_LABELS[model_key]
        frames.append(df)
    out = pd.concat(frames, ignore_index=True)
    out["model_order"] = out["model_key"].map({key: i for i, key in enumerate(MODEL_ORDER)})
    out["overall_score_100"] = as_numeric(out["overall_score_100"])
    out["gate_issue"] = out["gate_issue"].astype(bool)
    return out.sort_values(["model_order", "person_index"]).reset_index(drop=True)

judge_summary_all = load_judge_summary_all()
judge_summary_all.to_csv(REPORT_DIR / "step22_medgemma_judge_summary_all_models.csv", index=False, encoding="utf-8-sig")

judge_model_summary = (
    judge_summary_all
    .groupby(["model_key", "model_label_short"])
    .apply(lambda group: pd.Series({
        "records": len(group),
        "mean_score": group["overall_score_100"].mean(),
        "median_score": group["overall_score_100"].median(),
        "std_score": group["overall_score_100"].std(),
        "min_score": group["overall_score_100"].min(),
        "max_score": group["overall_score_100"].max(),
        "pass_count": int((group["decision"] == "pass").sum()),
        "review_needed_count": int((group["decision"] == "review_needed").sum()),
        "gate_issue_count": int(group["gate_issue"].sum()),
    }))
    .reset_index()
)
judge_model_summary["pass_rate_percent"] = judge_model_summary["pass_count"] / judge_model_summary["records"] * 100
judge_model_summary["gate_issue_rate_percent"] = judge_model_summary["gate_issue_count"] / judge_model_summary["records"] * 100
judge_model_summary["model_order"] = judge_model_summary["model_key"].map({key: i for i, key in enumerate(MODEL_ORDER)})
judge_model_summary = judge_model_summary.sort_values("model_order")
display(judge_model_summary.drop(columns="model_order"))
judge_model_summary.to_csv(REPORT_DIR / "step22_medgemma_model_summary.csv", index=False, encoding="utf-8-sig")

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
ordered_summary = judge_model_summary.set_index("model_key").loc[MODEL_ORDER].reset_index()
labels = ordered_summary["model_label_short"]
colors = [MODEL_COLORS[k] for k in ordered_summary["model_key"]]

axes[0, 0].bar(labels, ordered_summary["mean_score"], yerr=ordered_summary["std_score"], color=colors, capsize=4)
axes[0, 0].set_title("MedGemma Judge Mean Score by Model")
axes[0, 0].set_ylim(75, max(90, ordered_summary["mean_score"].max() + 4))
axes[0, 0].set_ylabel("Overall score / 100")
axes[0, 0].tick_params(axis="x", rotation=25)
plot_bar_labels(axes[0, 0], fmt="{:.2f}", pad=0.35)

stack_df = ordered_summary.set_index("model_label_short")[["pass_count", "review_needed_count"]]
stack_df.plot(kind="bar", stacked=True, ax=axes[0, 1], color=["#54A24B", "#E45756"])
axes[0, 1].set_title("MedGemma Decision Counts")
axes[0, 1].set_ylabel("Rows")
axes[0, 1].tick_params(axis="x", rotation=25)

axes[1, 0].bar(labels, ordered_summary["gate_issue_rate_percent"], color=colors)
axes[1, 0].set_title("Gate Issue Rate")
axes[1, 0].set_ylabel("Gate issue rows (%)")
axes[1, 0].tick_params(axis="x", rotation=25)
plot_bar_labels(axes[1, 0], fmt="{:.1f}", pad=0.5)

score_data = [judge_summary_all.loc[judge_summary_all["model_key"] == k, "overall_score_100"] for k in MODEL_ORDER]
axes[1, 1].boxplot(score_data, labels=[MODEL_LABELS[k] for k in MODEL_ORDER], patch_artist=True)
axes[1, 1].set_title("MedGemma Score Distribution")
axes[1, 1].set_ylabel("Overall score / 100")
axes[1, 1].tick_params(axis="x", rotation=25)

save_figure("step22_medgemma_judge_overview.png")
````

### Steps 16-18 Detailed Interpretation: What the Student Experiments Mean

The student experiments should be read as recommendation-model distillation, not as age-prediction training. The age prediction pipeline already produced `predicted_body_age_years`; the LoRA students learn how to turn a predicted body age plus profile features into structured recommendation text.

There are three student variants because they answer different experimental questions:

- **Student from Qwen:** the cleanest teacher-student fidelity test, because the teacher is Qwen2.5-7B and the student is Qwen2.5-0.5B.
- **Student from Llama:** a cross-family distillation test, where a Qwen student learns from Llama-generated responses.
- **Student from both:** a multi-teacher experiment, where the training set is larger and more diverse but also mixes teacher styles.

### Step 22 Detailed Interpretation: How to Read MedGemma Judge Results

The MedGemma judge layer is the strongest comparison layer because it reviews the actual generated recommendation text. Still, it should be reported carefully. A judge score is not a medical ground truth label; it is a structured external evaluation of response quality, safety boundaries, and profile-specific reasoning.

The useful comparisons are:

- **Mean score:** overall ranking across the full aligned sample.
- **Score distribution:** whether a model is consistently good or has unstable low-score cases.
- **Win count:** how often a model receives the highest score for an individual profile.
- **Pairwise deltas:** whether a model is systematically above or below the Qwen teacher on the same profile.
- **Criteria-level gaps:** which exact evaluation domains create the score difference.

## Step 27: Rule-Based Hard Safety Checks

This step adds the first final safety layer: deterministic hard safety checks. Unlike an LLM judge, this evaluator does not interpret the response holistically. It searches for specific safety boundaries and required context rules derived from the project's professional checklist.

The rule-based layer is useful because it is:

- **Deterministic:** the same response receives the same rule result every time.
- **Auditable:** each violation maps to a rule id, severity, category, source, and evidence string.
- **Strict on hard boundaries:** direct diagnosis, medication directives, nicotine replacement recommendations, chronological-age claims, and missing high-risk context can be flagged explicitly.
- **Complementary to MedGemma and GPT-5.5:** it catches checklist failures that a broad judge score may compress into a single number.

The important interpretation is that the rule-based gate is intentionally conservative. A lower rule-based gate pass rate does not necessarily mean the model is useless; it means many outputs need stricter safety context or checklist coverage before being treated as final recommendation text.


````python
# FINAL_SAFETY_REFERENCE_HUMAN_V1
# Step 27 setup: final safety-layer paths, fallback helpers, and file loading.
from pathlib import Path
import json
import math
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, HTML

PROJECT_ROOT = Path.cwd()
OUTPUT_DIR = PROJECT_ROOT / "model_outputs"
FIGURE_DIR = OUTPUT_DIR / "figures"
REPORT_DIR = OUTPUT_DIR / "reports"
RULE_SAFETY_DIR = OUTPUT_DIR / "rule_based_hard_safety_checks"
GPT55_DIR = PROJECT_ROOT / "evaluation_manual_gpt55"
GPT55_RESULTS_DIR = GPT55_DIR / "results"
HUMAN_REVIEW_DIR = GPT55_DIR / "human_review_20"
FINAL_JUDGE_DIR = PROJECT_ROOT / "final judge data"
FIGURE_DIR.mkdir(parents=True, exist_ok=True)
REPORT_DIR.mkdir(parents=True, exist_ok=True)

if "MODEL_LABELS" not in globals():
    MODEL_LABELS = {
        "teacher_qwen25_7b": "Teacher: Qwen2.5-7B",
        "teacher_llama31_8b": "Teacher: Llama 3.1 8B",
        "student_qwen_from_qwen": "Student: 0.5B from Qwen",
        "student_qwen_from_llama": "Student: 0.5B from Llama",
        "student_qwen_from_both": "Student: 0.5B from Both",
    }
if "MODEL_ORDER" not in globals():
    MODEL_ORDER = list(MODEL_LABELS.keys())
if "MODEL_COLORS" not in globals():
    MODEL_COLORS = {
        "teacher_qwen25_7b": "#4C78A8",
        "teacher_llama31_8b": "#F58518",
        "student_qwen_from_qwen": "#54A24B",
        "student_qwen_from_llama": "#B279A2",
        "student_qwen_from_both": "#E45756",
    }

MODEL_ID_TO_KEY = {
    "M1_teacher_Qwen2.5-7B": "teacher_qwen25_7b",
    "M2_teacher_Llama3.1-8B": "teacher_llama31_8b",
    "M3_student_Qwen2.5-0.5B_from_Qwen2.5-7B": "student_qwen_from_qwen",
    "M4_student_Qwen2.5-0.5B_from_Llama3.1-8B": "student_qwen_from_llama",
    "M5_student_Qwen2.5-0.5B_from_Qwen2.5-7B_plus_Llama3.1-8B": "student_qwen_from_both",
}

def read_csv_safe(path, **kwargs):
    path = Path(path)
    if not path.exists():
        raise FileNotFoundError(f"Missing expected CSV: {path}")
    with path.open("r", encoding="utf-8-sig", newline="") as handle:
        return pd.read_csv(handle, **kwargs)

def as_numeric(series):
    return pd.to_numeric(series, errors="coerce")

def clean_text(text):
    return str(text if pd.notna(text) else "").replace("\r\n", "\n").replace("\r", "\n").strip()

def save_figure(filename):
    path = FIGURE_DIR / filename
    plt.tight_layout()
    plt.savefig(path, dpi=180, bbox_inches="tight")
    plt.show()
    display(Markdown(f"Saved figure: `{path.relative_to(PROJECT_ROOT)}`"))
    return path

def plot_bar_labels(ax, fmt="{:.1f}", pad=0.5):
    for patch in ax.patches:
        height = patch.get_height()
        if np.isfinite(height):
            ax.text(
                patch.get_x() + patch.get_width() / 2,
                height + pad,
                fmt.format(height),
                ha="center",
                va="bottom",
                fontsize=8,
            )

def parse_issue_counts(issue_series):
    counts = {}
    for value in issue_series.fillna(""):
        for part in str(value).split(";"):
            part = part.strip()
            if not part:
                continue
            match = re.match(r"(.+?)\s*\((\d+)\)", part)
            if match:
                issue = match.group(1).strip()
                count = int(match.group(2))
            else:
                issue = part
                count = 1
            counts[issue] = counts.get(issue, 0) + count
    return pd.DataFrame([{"issue": k, "count": v} for k, v in counts.items()]).sort_values("count", ascending=False)

def load_medgemma_summary_all():
    dirs = {
        "teacher_qwen25_7b": FINAL_JUDGE_DIR / "01_teacher_Qwen2.5-7B",
        "teacher_llama31_8b": FINAL_JUDGE_DIR / "02_teacher_Llama_3.1_8B",
        "student_qwen_from_qwen": FINAL_JUDGE_DIR / "03_student_Qwen2.5-0.5B_from_Qwen2.5-7B",
        "student_qwen_from_llama": FINAL_JUDGE_DIR / "04_student_Qwen2.5-0.5B_from_Llama_3.1_8B",
        "student_qwen_from_both": FINAL_JUDGE_DIR / "05_student_Qwen2.5-0.5B_from_Qwen2.5-7B_plus_Llama_3.1_8B",
    }
    frames = []
    for model_key, directory in dirs.items():
        path = directory / "medgemma_judge_summary.csv"
        if path.exists():
            df = read_csv_safe(path)
            df["model_key"] = model_key
            df["model_label_short"] = MODEL_LABELS[model_key]
            frames.append(df)
    out = pd.concat(frames, ignore_index=True)
    out["overall_score_100"] = as_numeric(out["overall_score_100"])
    out["gate_issue"] = out["gate_issue"].astype(bool)
    return out

rule_model_summary = read_csv_safe(RULE_SAFETY_DIR / "rule_based_hard_safety_model_summary.csv")
rule_response_scores = read_csv_safe(RULE_SAFETY_DIR / "rule_based_hard_safety_response_scores.csv")
rule_summary = read_csv_safe(RULE_SAFETY_DIR / "rule_based_hard_safety_rule_summary.csv")
with (RULE_SAFETY_DIR / "rule_based_hard_safety_rulebook.json").open("r", encoding="utf-8-sig") as handle:
    rulebook = json.load(handle)

rule_model_summary["model_order"] = rule_model_summary["model_key"].map({key: i for i, key in enumerate(MODEL_ORDER)})
rule_model_summary = rule_model_summary.sort_values("model_order")
display(rule_model_summary)
display(Markdown(f"**Rulebook size:** {len(rulebook)} hard-safety and required-context rules."))
````

## Step 30: Final Triangulated Safety and Quality Comparison

This final step combines all evaluation layers into one interpretation:

1. **MedGemma full judge:** broad 2933-row LLM judge across all five models.
2. **Rule-based hard safety checker:** deterministic 2933-row checklist and safety-boundary audit.
3. **GPT-5.5 reference judge:** smaller 250-row blinded reference evaluation.
4. **Human review:** targeted 20-row high-risk audit.

These layers answer different questions. The final report should not collapse them into one fake accuracy number. Instead, it should say that the recommendation outputs were checked through multiple complementary lenses: broad LLM judging, deterministic safety rules, independent reference judging, and targeted human of difficult cases.


````python
# FINAL_SAFETY_REFERENCE_HUMAN_V1
# Step 30: build final triangulated comparison table.
medgemma_all = load_medgemma_summary_all()
medgemma_model_summary = (
    medgemma_all.groupby(["model_key", "model_label_short"])
    .apply(lambda g: pd.Series({
        "medgemma_rows": len(g),
        "medgemma_mean_score": g["overall_score_100"].mean(),
        "medgemma_pass_rate_percent": g["decision"].eq("pass").mean() * 100,
        "medgemma_gate_issue_rate_percent": g["gate_issue"].mean() * 100,
    }))
    .reset_index()
)

triangulated = (
    medgemma_model_summary
    .merge(rule_model_summary[[
        "model_key", "mean_rule_based_safety_score_100", "hard_policy_pass_rate_percent",
        "rule_based_gate_pass_rate_percent", "required_context_failure_rate_percent",
        "critical_violation_responses", "major_violation_responses",
    ]], on="model_key", how="left")
    .merge(gpt55_by_model[[
        "model_key", "pass_rate_percent", "average_medical_correctness", "average_safety_risk",
        "average_policy_compliance", "number_of_evaluated_rows",
    ]].rename(columns={
        "pass_rate_percent": "gpt55_pass_rate_percent",
        "average_medical_correctness": "gpt55_avg_medical_correctness",
        "average_safety_risk": "gpt55_avg_safety_risk",
        "average_policy_compliance": "gpt55_avg_policy_compliance",
        "number_of_evaluated_rows": "gpt55_rows",
    }), on="model_key", how="left")
    .merge(human_by_model[[
        "model_key", "reviewed_rows", "manual_pass_rate_percent", "critical_count", "major_count",
        "avg_manual_medical_correctness", "avg_manual_safety_risk", "avg_manual_policy_compliance",
    ]].rename(columns={
        "reviewed_rows": "human_targeted_rows",
        "manual_pass_rate_percent": "human_targeted_pass_rate_percent",
        "critical_count": "human_critical_count",
        "major_count": "human_major_count",
    }), on="model_key", how="left")
)
triangulated["model_order"] = triangulated["model_key"].map({key: i for i, key in enumerate(MODEL_ORDER)})
triangulated = triangulated.sort_values("model_order")
display(triangulated.drop(columns="model_order"))
triangulated.to_csv(REPORT_DIR / "step30_final_triangulated_safety_quality_table.csv", index=False, encoding="utf-8-sig")
````
